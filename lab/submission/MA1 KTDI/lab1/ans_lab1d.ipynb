# Lab 1d - Filtering and Sorting Data

Please use the [**Chipotle Sales dataset**](https://raw.githubusercontent.com/drshahizan/dataset/main/pandas/chipotle.tsv) for this exercise.

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
path = "https://raw.githubusercontent.com/drshahizan/dataset/main/pandas/chipotle.tsv"

chipo = pd.read_csv(path , sep = "\t")

chipo

```

### Step 4. How many products cost more than $10.00?

```python
chipo.item_price.dtype
#type(chipo.columns[5])
chipo.dtypes
```

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
products = len(chipo[chipo['item_price'] > 10.0])
products
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
chipo['priceEachItem'] = (chipo['item_price'] / chipo['quantity'])

print(chipo[['item_name' , 'priceEachItem']])
```

### Step 6. Sort by the name of the item

```python
sorted_chipo = chipo.sort_values(by='item_name', ascending=True)
sorted_chipo
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item = sorted_chipo.sort_values(by='item_price', ascending=False)
expitemloc = most_expensive_item.iloc[0]
Expitem = expitemloc['quantity']
print(Expitem)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie = (chipo['item_name'] == "Veggie Salad Bowl").sum()
print(veggie)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
cannedSoda = (chipo['item_name'] == "Canned Soda") & (chipo['quantity'] > 1 )
can = cannedSoda.sum()
print(can)
```
