#### Step 12. How many different occupations are in this dataset?

```python
unique_occupations = users['occupation'].unique()
print("Number of different occupations: ", len(unique_occupations))
```

```python
count_occupations = users['occupation'].value_counts().idxmax()
print("Most frequent occupation: ", count_occupations)
```
