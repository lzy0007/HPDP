#### Step 1. Import the necessary libraries

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

#### Step 2. Import the dataset from this address.

```python
path = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

#### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(path, delimiter='\t')

# Display the DataFrame
chipo
```

```python
chipo.info()
```

#### Step 4. How many products cost more than $10.00?

```python
def convert_float(price):
    return float(price[1:])

chipo['item_price'] = chipo['item_price'].apply(convert_float)

expensive_items = chipo[chipo['item_price'] > 10.00]

total_moreten = expensive_items.shape[0]

print("Number of products that cost more than $10.00:", total_moreten)
```

#### Step 5. What is the price of each item?

##### Print a data frame with only two columns item_name and item_price


```python
item_prices_list = chipo[['item_name', 'item_price']]

item_prices_list
```

#### Step 6. Sort by the name of the item

```python
sorted_by_name= item_prices_list.sort_values(by='item_name')

sorted_by_name
```

#### Step 7. What was the quantity of the most expensive item ordered?

```python
quantity_of_most_expensive_item = chipo.loc[chipo['item_price'].idxmax(), 'quantity']

print("Quantity of the most expensive item ordered:", quantity_of_most_expensive_item)
```

#### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
num_veggie_salad_orders = chipo.loc[chipo['item_name'] == 'Veggie Salad Bowl', 'quantity'].sum()

print("Number of times Veggie Salad Bowl was ordered:", num_veggie_salad_orders)
```

#### Step 9. How many times did someone order more than one Canned Soda?###

```python
num_multiple_canned_soda_orders = ((chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)).sum()

print("Number of times someone ordered more than one Canned Soda:", num_multiple_canned_soda_orders)
```
