###Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [chipotle.tsv](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv)

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')
chipo
```

### Step 4. How many products cost more than $10.00?

We change the 'item_price' column from a string to a float type so that we can filtered out the products cost more than $10.00

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '')
chipo['item_price'] = chipo['item_price'].astype(float)
ip = np.count_nonzero(chipo['item_price'] > 10.00)

print("There is {} quantity of products that are more than $10.00.".format(ip))
```

### Step 5. What is the price of each item?
print a data frame with only two columns item_name and item_price

```python
chipo.groupby('item_name')[['item_price']].min().reset_index()
```

### Step 6. Sort by the name of the item



```python
sortitem = chipo
sortitem.sort_values(by='item_name').reset_index()
```

### Step 7. What was the quantity of the most expensive item ordered?

First of all, from the table below we can see there is 3 items that have the highest price of all from the list and its quantity

```python
chipo.groupby('item_name')[['item_price']].min().reset_index()
item_summary = chipo.groupby('item_name').agg({'item_price': 'min', 'quantity': 'sum'}).reset_index()
item_summary
```

We take the 3 most expensive items then we sum up the quantity of the order

```python
exp = chipo[chipo['item_name'].isin(['Barbacoa Salad Bowl', 'Carnitas Salad Bowl', 'Steak Salad Bowl'])]
exp_chipo = exp['quantity'].sum()
print("There are {} numbers of the 3 most expensive items ".format(exp_chipo))
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
vsb = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
veggie = vsb.quantity.sum()
print("There are {} of times a Veggie Salad Bowl ordered by customer".format(veggie))
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
cSoda = chipo[chipo['item_name'] == 'Canned Soda']
cs = len(cSoda[cSoda['quantity'] > 1]) #filtered order more than one canned soda.
print("There are {} number of times someone order more than one Canned Soda." .format(cs) )
```
